```python
def categorizar_tomate_media (media):
  if media >= 40 and media <= 70:
    return "tomate medio"
  if media < 40:
    return "tomate pequeno"
  else:
    return "tomatão"
```

```python
df_csv["categoria_tomate"] = df_csv["Average"].apply(categorizar_tomate_media)
df_csv.head()
```

```python
df_chess = df_excel
df_chess.head()
```

```python
#grafico de dispersão


plt.scatter(df_chess["black_rating"],df_chess["white_rating"])
plt.xlabel("Black")
plt.ylabel("White")
plt.title("PArtidas de peças pretaz x peças brancas")
plt.show()
```

```python
#grafico de barras para analisar um volume de dados dado categorias

plt.bar(df_tomato["categoria_tomate"],df_tomato["Average"])
plt.xlabel("Categoria de tomates")
plt.ylabel("Média")
plt.title("Média de tomates por categoria")
plt.show ()
```

```python
#cada fatia é a distribuição dos dados em relação ao total -(proporcionalidade)- pizza
df_chess.groupby(["victory_status"]).mean().plot(kind = "pie", y='turns', autopct="%1.0f%%")
plt.title("Média de partidas dentro de estatus de vitória")
```

*   pizza = proporções
*   pontos(dispersão)=como os dados estão se comportando
*   linhas = tendência(tempo)
*   barra = comparação

```python
df_diabetes = pd.read_csv("/content/drive/MyDrive/Pós Tech - Data Analytics/Fase 2/bases/diabetes.csv", sep = ",")
df_diabetes.head()
```

```python
#separar em treino e teste
from sklearn.model_selection import train_test_split
```

```python
x = df_diabetes.drop(["Class variable"],axis = 1)
```

```python
y = df_diabetes["Class variable"]
```

```python
#variavel target, ela me da exatamente a informação que eu quero chegar (se a pessoa tem ou não diabetes)
y
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size = 0.3)
```

```python
len (x_train)
```

```python
df_diabetes.shape
```

```python
len (x_test)
```

```python
#classificação(caracteristicas próximas), o algoritmo funciona analisando a distancia dos dados, o que se assemelha ele agrupa
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors =3)
```

```python
knn.fit(x_train,y_train)
```

```python
accuracy = knn.score(x_test,y_test)
```
